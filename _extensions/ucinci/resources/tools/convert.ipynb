```python
#!/usr/bin/env python3
# -*- coding: utf-8 -*-
"""
Created on Mon Jun 23 07:49:05 2025

@author: paytone
"""

def convert_all_cwd_ppt2md():
    import os
    import pypandoc as ppd

    # src = os.getcwd()
    # for dir, dirs, files in os.walk(src):
    #     files = [os.path.join(dir, f) for f in files if not f.endswith(('.py', '.pptx'))]

    # if files
    # for file in files:
    #     fnom = file.split(os.sep)[-1].split('.')[0]
    #     ppd.convert_file(file, 'md', outputfile=fnom + ".txt")

    #%%

    from pptx2md import convert, ConversionConfig
    from pathlib import Path

    src = os.getcwd()

    files = [os.path.join(src, f) for f in os.listdir(src) if f.endswith(('.pptx'))]
    #%%
    for file in files:
        fnom = file.split(os.sep)[-1].split('.')[0]
        # Basic usage
        convert(
            ConversionConfig(
                pptx_path=Path(file),
                output_path=Path(fnom+".qmd"),
                output_dir=Path(os.sep.join(file.split(os.sep)[0:-1])),
                image_dir=Path('img'),
                is_qmd=True
            )
        )
```

```python
convert_all_cwd_ppt2md()
```
